**Carga del archivo xml**

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Carga de un archivo xml con indice 
df1 = pd.read_excel("Detalle precios y productos fabricados 2022.xlsx", index_col=5)
df1
```

```python
#Verificamos información del DataFrame
df.info()
```

**Aplicación de filtros**


```python
#Filtro por objeto: De la columna señalada, el filtro solo mostrara las filas que cumplan con la condición especifica
#Variable: NOMBRE_VENDEDOR (Variable que define el nombre de quien que vendio determinado el artículo en la tienda)
filtro1 = df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro1
```

```python
#Filtro por filas 
#Los ":" significa que selecciona todas las columnas
filtro2 = df.iloc[1:4,:] #Filas consecutivas: Selecciona la fila 1, 2 y 3 (Siempre excluye la ultima fila [4])
#filtro2= df.loc[[1,4],:] #Filas no consecutivas: Selecciona unicamente las filas 1 y 4
filtro2
```

```python
#Filtro por columnas 
#Los ":" significa que selecciona todas las filas
#filtro3 = df.iloc[:, 1:4] #Columnas consecutivas: Selecciona la columana 1, 2 y 3 
filtro3 = df.iloc[:, [2,4,5]] #Columnas consecutivas: Selecciona unicamente las columnas especificada, 
#en el caso de este ejemplo da como resultado el nombre del vendedor, el artículo que vendio y la cantidad
filtro3
```

```python
#Paso 2: Aplicamos filtros de filas vs columnas (el archivo debe tener un indice)
#df.loc[filas, columnas] -> La función .loc[] se utiliza para seleccionar filas y columnas del DataFrame por etiquetas
# (es decir, usando los valores de los índices y nombres de las columnas)
filtro4 = df1.loc[["TUE00244", "CS12F1040A00"], ["CANT", "NOMBRE_VENDEDOR"]]
#En este caso mostramos la clave del producto, la cantidad vendida y quien lo vendio
filtro4
```

```python
#Filtro por cabecera: devuelve las primeras 5 filas del dataFrame, ayuda a ver la estructura general 
filtro5 = df.head()
filtro5
```

```python
#Filtro por cola: muestra las ultimas 5 filas del DataFrame, dentro de los () especificase puede especificar el numero de filas que queremos que aparezcan
filtro6 = df.tail()
filtro6
```

```python
#Filtros por comparacion
filtro7 = df[df["CANT"] > 50] #selecciona las filas donde el valor en la columna CANT sea mayor a 50
filtro7
```

```python
#Filtro Y: agregamos el conector logico & para que cumpla mas de una condición, debe cumplir ambas a la vez
filtro8 = df[(df["SUBTOTAL_PARTIDA"] < 10000) & (df["CVE_ART"] == "CAS00083")]
filtro8
```

```python
#Filtro o: el conector logico | hace que se puedan cumplir mas de una condicion, pero no necesariamente ambas a la vez, puede ser una u otra
filtro9 = df[(df["SUBTOTAL_PARTIDA"] < 10000) | (df["CVE_ART"] == "CAS00083")]
filtro9
```

```python
#Filtro not: selecciona todas las filas del DataFrame excepto aquellas donde el valor de la columna "NOMBRE_VENDEDOR" sea "LETICIA RAMIREZ HERNANDEZ".
filtro10=df[~(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro10
```

**Creación de archivos CSV**
