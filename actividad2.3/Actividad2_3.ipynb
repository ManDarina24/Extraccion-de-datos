**Carga del archivo xml**

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Milan.csv')
df
```

```python
#Verificamos información del DataFrame
df.info()
```

**Aplicación de filtros**


```python
#Los registros de 5 host name diferentes 
filtro1 = df[(df["host_name"] == "Francesca") | (df["host_name"] == "Jian Luca") | (df["host_name"] == "Jeremy") | (df["host_name"] == "Laura") | (df["host_name"] == "Luciano") ]
filtro1
```

```python
#Unidos antes del año 2019
filtro2 = df[df["host_since"] < "2019-01-01"]
filtro2
```

```python
#Responden a mas tardar en un dia 
filtro3 = df[
    (df["host_response_time"] == "within a day")]
filtro3
```

```python
#Al menos 5 accommodates
filtro4 = df[df["accommodates"] >= 5]
filtro4
```

```python
#Tipos de cuarto Entire home y private room
filtro5 = df[
    (df["room_type"] == "Entire home/apt") |
    (df["room_type"] == "Private room") 
]
filtro5
```

```python
#A lo mucho cuentan con 4 camas 
filtro6 = df[df["beds"] <= 4]
filtro6
```

```python
#Los registros que tengan un baño o sean superhost pero con la condicion de que todos sean instant_bookable
filtro7 = df[
    ((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) & 
    (df["instant_bookable"] == "t")
]
filtro7
```

```python
filtro8 = df[(df["review_scores_cleanliness"] < 3) | (df["review_scores_rating"] > 3)  ]
filtro8
```

**Creación de archivos CSV**

```python
#Convertir archivo filtrado a csv
filtro1.to_csv("Filtro1.csv")
filtro2.to_csv("Filtro2.csv")
filtro3.to_csv("Filtro3.csv")
filtro4.to_csv("Filtro4.csv")
filtro5.to_csv("Filtro5.csv")
filtro6.to_csv("Filtro6.csv")
filtro7.to_csv("Filtro7.csv")
filtro8.to_csv("Filtro8.csv")
```
